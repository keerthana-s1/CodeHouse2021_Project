Sentiment Analysis


```python
polarity = []
subjectivity = []
for review in reviews:
   b1=TextBlob(review)
   #print(b1.sentiment)
   polarity = polarity + [b1.sentiment.polarity]
   subjectivity = subjectivity + [b1.sentiment.subjectivity]
```

```python
polarity_cisco = []
subjectivity_cisco = []
for review in reviews_cisco:
   b1=TextBlob(review)
   #print(b1.sentiment)
   polarity_cisco = polarity_cisco + [b1.sentiment.polarity]
   subjectivity_cisco = subjectivity_cisco + [b1.sentiment.subjectivity]
```

```python
review_tokens = []
for review in raw_reviews:
   words = nltk.word_tokenize(review)
   review_tokens+=nltk.pos_tag(words)
```

```python
adjectives=dict();
comment_words=''
for token in review_tokens:
    if token[1]=='JJ':
     if TextBlob(token[0]).sentiment.polarity>0:
       #print(token[0])
       if token[0] not in adjectives:
        adjectives[token[0]] = 1
        comment_words+=token[0]+" " 
       else:
        adjectives[token[0]] = adjectives[token[0]] + 1
adjectives
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords=set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800,background_color ='white',stopwords = stopwords,min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
